# Regression Learning

L**inear regression**

It is a statistical method used to model the relationship between a dependent variable and one or more independent variables. It is a linear approach to modeling the relationship between a scalar response (or dependent variable) and one or more explanatory variables (or independent variables). The goal of linear regression is to find the **best-fitting straight line through the data points**. The best-fitting line is called a regression line. 

The line can be represented mathematically using an equation of the form Y = a + bX where Y is the dependent variable, X is the independent variable, a is the y-intercept (the point at which the line crosses the y-axis) and b is the slope of the line. This equation can be used to make predictions about the value of the dependent variable based on the value of the independent variable.



Pros: fast, no tuning required, highly interpretable, well-understood

Cons: unlikely to produce the best predictive accuracy (presumes a linear relationship between the features and response)

# Decision Tree Learning

**Decision Tree Algorithm**

A decision tree is a flowchart-like tree structure that is used to represent a series of decisions or choices and their possible outcomes. It is a type of supervised machine learning algorithm that is used for both classification and regression tasks. The decision tree algorithm creates a tree-like model of decisions and their possible consequences, with each internal node of the tree representing a "test" on an attribute (e.g. whether a customer's income is above a certain threshold), each branch represents the outcome of the test, and each leaf node represents a class label (e.g. whether the customer is likely to buy a product or not).

decision-tree.png

The decision tree algorithm starts at the root node, which represents the entire sample, and splits the sample into subsets based on the values of the input features. The algorithm then recursively applies the same process to each subset, creating new internal nodes and branches, until it reaches a leaf node which represents a class label.

The algorithm uses a measure of impurity, such as entropy or Gini impurity, to determine which feature to split on at each internal node. The feature that results in the highest reduction in impurity is chosen as the split feature. The process is repeated on each child node, and this continues until a stopping criterion is met, such as a maximum depth of the tree or a minimum number of samples in a leaf node.

Decision trees are easy to understand and interpret, and can be used for both classification and regression tasks. However, they can also be prone to overfitting, especially when the tree becomes too deep, so it's important to prune the tree or use techniques like bagging or random forest to mitigate overfitting.

# Random Forest Learning

**Random Forest** is an ensemble machine learning algorithm that is used for both classification and regression tasks. It is a type of decision tree algorithm that is composed of multiple decision trees. The idea behind the random forest algorithm is to combine multiple decision trees in order to improve the overall performance and stability of the model.

The algorithm works by creating multiple decision trees using a technique called [bootstrap aggregating](https://en.wikipedia.org/wiki/Bootstrap_aggregating) or "bagging" on different subsets of the data. Each decision tree is trained on a different subset of the data, which is randomly sampled with replacement. The subsets of the data are created by drawing random samples from the original dataset with replacement, so the same data point can be included multiple times in the same subset.

RandomForest.png

In addition to bagging, Random Forest also uses a technique called random feature selection, which means that at each split point in the decision tree, a random subset of the input features is considered for the split, instead of all the features. This helps to reduce the correlation between the different decision trees, and therefore increase the diversity among the trees in the forest.

Once all the decision trees are trained, the predictions of the random forest are made by averaging the predictions of all the decision trees. For classification tasks, the majority vote of the decision trees is used to make the final prediction. Random Forest is considered to be a very powerful algorithm and it is often used in industry **to solve complex problems because it provides a good trade-off between bias and variance** and it also reduces overfitting.

# Support Vector Machine (SVM)

**Support Vector Machines** (SVMs) is a type of supervised machine learning algorithm that can be used for both classification and regression tasks. The main idea behind SVMs is to find a decision boundary or a hyperplane that separates the data into different classes. The hyperplane that maximizes the margin, which is the distance between the decision boundary and the closest data points of each class, is chosen as the best decision boundary.

support-vector-machine.png

The data points that are closest to the decision boundary are called support vectors. These points have the greatest impact on the position of the decision boundary. The decision boundary is chosen in such a way that it is equidistant from the closest points of each class, which are called support vectors.

In the case of non-linearly separable data, SVMs can still be used by applying a technique called the **kernel trick**. The kernel trick maps the input data into a higher dimensional space where it becomes linearly separable. The most commonly used kernel functions are linear, polynomial, and radial basis functions (RBF).

SVMs are considered to be very powerful algorithms, particularly in high-dimensional spaces, and are often used in **various applications such as text classification, image classification, and bioinformatics**. However, SVMs can be sensitive to the choice of kernel function and the parameters of the model, which can affect the performance of the model.

# Neural Network & Deep Learning

A **neural network** (ANN) is a type of machine learning algorithm that is inspired by the structure and function of the human brain. Neural networks are a set of algorithms, modeled loosely after the human brain, that are designed to recognize patterns. They interpret sensory data through a kind of machine perception, labeling or clustering raw input. The patterns they recognize are numerical, contained in vectors, into which all real-world data, be it images, sound, text or time series, must be translated.

Neural networks consist of layers of interconnected nodes, called artificial neurons, which are inspired by the structure and function of biological neurons in the human brain. Each neuron receives input from other neurons, performs a simple mathematical operation on the input, and passes the result to other neurons in the next layer. The input layer receives the raw input data, and the output layer produces the final output. The layers in between the input and output layers are called hidden layers.

The most common type of neural network is the **feedforward neural network**, in which the data flows in one direction, from input to output, through the layers of neurons. There are also other types of neural networks such as **recurrent neural networks** (RNN) and **convolutional neural networks** (CNN) that are used for specific tasks.

Neural networks are able to learn and improve their performance through a process called training, which involves adjusting the weights of the connections between the neurons based on the input data and the desired output. Training a neural network involves iteratively adjusting the weights of the connections between the neurons based on the input data and the desired output, until the network is able to produce the desired output for a given input.
Neural networks are powerful algorithms that can be used for a wide variety of tasks, such as **image recognition, natural language processing, and time series forecasting**. However, they can be computationally expensive and require large amounts of data for training.

**Long Short Term Memory (LSTM)**

LSTM is an artificial recurrent neural network (RNN) architecture used in the field of deep learning. Unlike standard feed-forward neural networks, LSTM has feedback connections. It can not only process single data points (such as images), but also entire sequences of data (such as speech or video). LSTM is widely used for the problems of **sequence prediction** and been very effective.

**LSTM in scikit-learn**

In this application, we used the LSTM network to predict the closing stock price using the past 60-day stock price.

```python
# plotting date vs the close market stock price
stock_data.plot(y='Close',color="red")
 
# extract only top 60 rows to make the plot a little clearer
new_data = stock_data.tail(60)
 
# plotting date vs the close  market stock price
new_data.plot(y='Close',color="green")

plt.show()
```

```python
# import all the necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, LSTM
import math
from sklearn.preprocessing import MinMaxScaler
```

To make our study easier we will only consider the **closing market price** and predict the closing market price using Scikit-learn. The whole train data preparation is shown in the steps below.

```python
# 1. Filter out the closing market price data
close_data = stock_data.filter(['Close'])
 
# 2. Convert the data into array for easy evaluation
dataset = close_data.values
 
# 3. Scale/Normalize the data to make all values between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)
 
# 4. Creating training data size : 70% of the data
training_data_len = math.ceil(len(dataset) *.7)
train_data = scaled_data[0:training_data_len  , : ]
 
# 5. Separating the data into x and y data
x_train_data=[]
y_train_data =[]
for i in range(60,len(train_data)):
    x_train_data=list(x_train_data)
    y_train_data=list(y_train_data)
    x_train_data.append(train_data[i-60:i,0])
    y_train_data.append(train_data[i,0])
 
    # 6. Converting the training x and y values to numpy arrays
    x_train_data1, y_train_data1 = np.array(x_train_data), np.array(y_train_data)
 
    # 7. Reshaping training x and y data to make the calculations easier
    x_train_data2 = np.reshape(x_train_data1, (x_train_data1.shape[0],x_train_data1.shape[1],1))
```

```python
training_data_len
```

Here we create a data set to train the data that contains the closing price of 60 days ( 60 data points) so that we could do the prediction for the 61st closing price.

Now the x_train data set will contain a total of 60 values, the first column will contain from the index of 0 to 59 and the second column from the index of 1 to 60, and so on.

The y_train data set will contain the 61st value at its first column located at index 60 and for the second column, it will contain the 62nd value located at index 61 and so on.

Converting both the independent and dependent train data set as x_train_data and y_train_data respectively, into the NumPy arrays so that they can be used to train the LSTM model.

Also, as the LSTM model is expecting the data in 3-dimensional data set, using reshape() function we will reshape the data in the form of 3-dimension.

**Building LSTM Model**

The LSTM model will have two LSTM layers with 50 neurons and two Dense layers, one with 25 neurons and the other with one neuron.

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True,input_shape=(x_train_data2.shape[1],1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=25))
model.add(Dense(units=1))
```

**Compiling the Model**

The LSTM model is compiled using the mean squared error (MSE) loss function and the adam optimizer.

Using the fit() function which is another name for train, we are training the data sets. Here, batch_size is the total number of training examples present in the single batch, and epochs are the number of iterations when an entire data set is passed forward and backward through the neural network.

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train_data2, y_train_data1, batch_size=1, epochs=1)
```

# FAQs

1. Can I use cross validation (CV) for time series prediction?

Yes, cross validation (CV) can be used for time series prediction, but it requires special considerations compared to regular CV. In time series data, the samples are often correlated and the order of the samples is important. Therefore, the traditional k-fold CV technique may not be appropriate as it randomly splits the data into k folds, which can lead to a loss of temporal structure and dependencies in the data. To address this, various modifications to k-fold CV have been proposed for time series, such as forward chaining, backward chaining, and rolling window cross validation. These methods aim to preserve the temporal structure of the data while performing the cross-validation.

2. What is rolling window cross validation?

Rolling window cross validation is a technique used for time series prediction that aims to preserve the temporal structure of the data while performing cross-validation. In this method, the time series data is divided into overlapping windows, and each window is used as the validation set, while the remaining data is used as the training set. The process is then repeated, with each window being used as the validation set once. This allows the model to be trained on a longer time period and validated on a shorter time period, while still preserving the temporal structure of the data. The results of each validation step can be combined to obtain a final estimate of the model performance. Rolling window cross validation can be a good choice when the goal is to evaluate the model's ability to make predictions for future time points based on past observations.

```python
# Here is an example of how you can implement rolling window cross validation in Python:

import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error

def rolling_window_cv(model, X, y, window_size):
    tscv = TimeSeriesSplit(n_splits=window_size)
    mse = []
    for train_index, test_index in tscv.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mse.append(mean_squared_error(y_test, y_pred))
    return np.mean(mse)
```

In this example, model is the machine learning model you want to use for time series prediction, X and y are the feature matrix and target vector, respectively. The rolling_window_cv function takes the model, feature matrix, target vector, and window size as inputs. The TimeSeriesSplit class from sklearn.model_selection is used to perform the rolling window cross-validation, and mean_squared_error from sklearn.metrics is used to evaluate the model performance. The function returns the average mean squared error across all windows.

3. What is LSTM and its use for stock prediction?

LSTM stands for Long-Short-Term Memory, a type of Recurrent Neural Network (RNN) architecture that is particularly well-suited for processing sequential data like time series. In stock prediction, LSTM can be used to analyze historical stock prices and other financial data, and make predictions about future trends. The LSTM algorithm can learn from past patterns in the data, and use that knowledge to make informed predictions about future stock prices. This approach has shown to be effective in stock price forecasting, and is widely used in finance and investment.

4. Can I use multiple features as input to LSTM?

Yes, you can use multiple features as input to an LSTM network. In fact, it is common to use multiple features in stock prediction tasks to improve the accuracy of the model. 5. What is sliding window approach?

The sliding window approach is a method used in sequence data processing to extract a set of contiguous data points from a larger sequence. In this approach, a fixed-sized window or sub-sequence is moved step-by-step along the sequence, capturing a new set of data points at each step. The size of the window and the step size can be adjusted to control the granularity of the extracted data.

The sliding window approach is commonly used in time series analysis, where it is used to create inputs and outputs for machine learning models such as LSTMs, ARIMA, etc. The inputs represent a fixed number of past time steps, and the outputs represent the value at the next time step. This way, the sliding window approach enables the models to learn the temporal dependencies between the time steps and make predictions accordingly.

6. What is ARIMA?

**ARIMA** (AutoRegressive Integrated Moving Average) is a statistical model used for time series forecasting. It combines three components: an autoregression (AR) model, an integration (I) model, and a moving average (MA) model, to predict future values based on past values in a time series.

The autoregression component models the relationship between an observation and a number of lagged observations. The integration component is used to remove the non-stationarity from the time series, and the moving average component models the error term in the time series.

ARIMA models are particularly useful for time series that are stationary, meaning that their mean and variance do not change over time. The model is fit to the time series data, and its parameters are optimized to minimize the prediction error.

Once the ARIMA model is fit to the data, it can be used to make predictions for future time steps. ARIMA models are widely used in finance, economics, and other fields where time series forecasting is important.
